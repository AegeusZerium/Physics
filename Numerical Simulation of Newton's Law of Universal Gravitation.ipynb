## Newton's law of universal gravitation

**Newton's law of universal gravitation** states that every particle attracts every other particle in the universe with a force $F_g$ which is directly proportional to the product of their masses, $F_g \propto m_1 m_2$, and inversely proportional to the square of the distance between their centers, $F_g \propto \frac{1}{r^2}$.
<br>
The equation for universal gravitation thus takes the form,
<br><br>
$$F_g = G{\frac{m_1 m_2}{r^2}}$$
<br>
where $F_g$ is the gravitational force acting between two objects, $m_1$ and $m_2$ are the masses , $10^{−11}$, of the objects, $r$ is the distance between the centers of their masses, and $G \sim 6.674 \cdot$ $10^−11$ $m^{3}kg^{−1}s^{−2}$ is the **gravitational constant** [**[2]**](https://en.wikipedia.org/wiki/Newton%27s_law_of_universal_gravitation).

```python
# Scientific libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 1
M = 1
m = 0.1
r = np.linspace(0.1, 10, 100)

# Gravitational Force
F_g = G*M*m/r**2

# Visualization
plt.figure(figsize=(5,5))
plt.plot(r, F_g)
plt.xlabel(r"$r$")
plt.ylabel(r"$F_g$")
plt.title("Newton's inverse square law")
plt.grid(True)
plt.show()
```

The above graph is an illustration of Newton's **inverse square law**.

## Vector form

Newton's law of universal gravitation can be written as a **vector equation** to account for the direction of the gravitational force as well as its magnitude,
<br><br>
$$\textbf{F}_{12} = G \frac{m_1 m_2}{|\textbf{r}_{12}|^3}|\textbf{r}_{12}|\hat{\textbf{r}}_{12}$$
<br>
where, 
<br>
- $\textbf{F}_{12}$ is the force applied on object $1$ exerted by object $2$,
- $G$ is the gravitational constant,
- $m_1$ and $m_2$ are respectively the masses of objects $1$ and $2$,
- $\left|\textbf{r}_{12}\right| = \left|\textbf{r}_2 − \textbf{r}_1\right|$ is the distance between objects $1$ and $2$, and
- ${\displaystyle \mathbf {\hat {r}} _{12}\ {\stackrel {\mathrm {def} }{=}}\ {\frac {\mathbf {r} _{2}-\mathbf {r} _{1}}{\vert \mathbf {r} _{2}-\mathbf {r} _{1}\vert }}}$ is the **unit vector** from object $1$ to $2$.

It can be seen that the vector form of the equation is the same as the scalar form given earlier, except that $\textbf{F}$ is now a vector quantity, and the right hand side is multiplied by the appropriate unit vector. Also, it can be seen that $\textbf{F}_{12} = -\textbf{F}_{21}$ [**[3]**](https://en.wikipedia.org/wiki/Newton%27s_law_of_universal_gravitation). 

## Euler method

**Euler's method** is a numerical method to solve first order first degree differential equation with a given initial value. It is the most basic explicit method for numerical integration of ordinary differential equations and is the simplest **Runge–Kutta method** [**[4]**](https://en.wikipedia.org/wiki/Euler_method).

```python
def EulerMethod(h, y0):
    y=np.zeros(5)
    for i in range(0, 4):
        y[0] = y0
        y[i+1] = y[i] + h*y[i] 
    return y

x = np.arange(0, 5)
plt.figure(figsize=(5,5))
plt.plot(x, np.exp(x), label="Exact solution")
plt.plot(x, EulerMethod(1, 1), label="Euler method")
plt.title(r"Illustration of numerical integration for the equation $y' = y, y_0 = 1$")
plt.grid(True)
plt.legend()
plt.show()
```

Euler's method is written in the following form,
<br><br>
$$y_{n+1} = y_n + h \cdot f(x_n, y_n)$$
<br>
where,
<br>
- $y_{n+1}$ is the estimated solution value,
<br>
- $y_n$ is the current value,
<br>
- $h$ is the step size,
<br>
- $f(x_n, y_n)$ is the value of the derivative at the starting point $x_n$ and $y_n$.

## Numerical simulation

The modern method consists of numerical integration in 3-dimensional space. One starts with a high accuracy value for the position $(x, y, z)$ and the velocity $(v_x, v_y, v_z)$ for each of the bodies involved. When also the mass of each body is known, the acceleration $(a_x, a_y, a_z)$ can be calculated from **Newton's Law of Gravitation**. Each body attracts each other body, the total acceleration being the sum of all these attractions. Next one chooses a small time-step $\Delta t$ and applies **Newton's Second Law of Motion**. The acceleration multiplied with $\Delta t$ gives a correction to the velocity. The velocity multiplied with Δt gives a correction to the position. This procedure is repeated for all other bodies.

The result is a new value for position and velocity for all bodies. Then, using these new values one starts over the whole calculation for the next time-step $\Delta t$. Repeating this procedure often enough, and one ends up with a description of the positions of all bodies over time.

## Two-body problem

In an N-body problem the equations of motion would be,
<br><br>
$$\ddot{\textbf{r}}_i = - G \sum_{j \neq i}^N \frac{m_j \left(\textbf{r}_i - \textbf{r}_j \right)}{\left|\textbf{r}_j - \textbf{r}_i \right|^3}$$
<br>
We have $\textbf{F}_{12}$ is the force on mass $1$ due to its interactions with mass $2$, 
<br><br>
$$\textbf{F}_{12} = m_1 \ddot{\textbf{r}}_1 = m_1 G \frac{m_2}{|\textbf{r}_{12}|^3}|\textbf{r}_{12}|\hat{\textbf{r}}_{12}$$
<br>
and $\textbf{F}_{21}$ is the force on mass $2$ due to its interactions with mass $1$, 
<br><br>
$$\textbf{F}_{21} = m_2 \ddot{\textbf{r}}_2 = - m_2 G \frac{m_1}{|\textbf{r}_{12}|^3}|\textbf{r}_{12}|\hat{\textbf{r}}_{12}$$
<br>
thus, the mathematical statement of the two-body problem can be given in terms of the Newtonian equations of motion as,
<br><br>
$$\begin{align*}
&\ddot{\textbf{r}}_1 = \frac{G m_2}{|\textbf{r}_{12}|^3}|\textbf{r}_{12}|\hat{\textbf{r}}_{12} \\
&\ddot{\textbf{r}}_2 = - \frac{G m_1}{|\textbf{r}_{12}|^3}|\textbf{r}_{12}|\hat{\textbf{r}}_{12}
\end{align*}$$
<br>
Using Euler's method, the solution to our system of differential equations would then be, for object $1$,
<br><br>
$$\dot{\textbf{r}}_1 \left(t + \Delta t\right) = \dot{\textbf{r}}_{1} \left(t\right) + \Delta t \cdot \ddot{\textbf{r}}_1 \left(t\right)$$
<br>
$$\textbf{r}_1 \left(t + \Delta t\right) = \textbf{r}_{1} \left(t\right) + \Delta t \cdot \dot{\textbf{r}}_1 \left(t\right)$$
<br>
for object $2$,
<br>
$$\dot{\textbf{r}}_2 \left(t + \Delta t\right) = \dot{\textbf{r}}_{2} \left(t\right) + \Delta t \cdot \ddot{\textbf{r}}_2 \left(t\right)$$
<br>
$$\textbf{r}_2 \left(t + \Delta t\right) = \textbf{r}_{2} \left(t\right) + \Delta t \cdot \dot{\textbf{r}}_2 \left(t\right)$$

## Implementation

We will simulate the orbit of Earth around the, stationary (which is in reality not stationary), Sun using the parameters of the former (Earth) relative to the latter (Sun),

| Parameters | Earth | Sun |
| --- | --- | --- |
| **Mass** | $5.972 \cdot 10^{24} \space kg$ | $1.9891 \cdot 10^{30} \space kg$ | 
| **Distance** | $1.5 \cdot 10^{11} \space m$ | $-$ |
| **Velocity** | $3 \cdot 10^5 \space m/s$ | $-$ |

```python
# Storing coordinate data
x_11list = []
x_12list = []

x_21list = []
x_22list = []

# Storing velocity data of the Earth
v_21list = []
v_22list = []

# Storing Force 21 (Force applied by the Sun on the Earth)
Fx_21list = []
Fx_22list = []

# Constants
G = 6.67408e-11 # m^3 kg^-1 s^-2
t = 0.0 # s
dt = 0.01*24*60*60 # s

# Sun Parameters
# mass
m_1 = 1.989e30 # kg
# position
x_11 = 0
x_12 = 0
x_13 = 0
# velocity
v_11 = 0
v_12 = 0
v_13 = 0

# Earth Parameters
# mass
m_2 = 5.972e24 # kg
# position
x_21 = 1.5e11 # m
x_22 = 0
x_23 = 0
# velocity
v_21 = 0
v_22 = 30000 # m/s
v_23 = 0

while t < 380*24*60*60:
    
    # Distance
    r_12 = np.sqrt((x_21-x_11)**2 + (x_22-x_12)**2 + (v_23-v_13)**2) # m
    
    # Newton's Second Law of Motion
    # Force 12
    Fx_11 = (G*m_1*m_2*(x_21-x_11))/r_12**3
    Fx_12 = (G*m_1*m_2*(x_22-x_12))/r_12**3
    Fx_13 = (G*m_1*m_2*(x_23-x_13))/r_12**3
        
    # Force 21
    Fx_21 = -(G*m_1*m_2*(x_21-x_11))/r_12**3
    Fx_22 = -(G*m_1*m_2*(x_22-x_12))/r_12**3
    Fx_23 = -(G*m_1*m_2*(x_23-x_13))/r_12**3
    
    # Euler Method
    # Sun
    v_11 += (Fx_11*dt)/m_1
    v_12 += (Fx_12*dt)/m_1
    v_13 += (Fx_13*dt)/m_1
    
    x_11 += v_11*dt
    x_12 += v_12*dt
    x_13 += v_13*dt
    
    # Earth
    v_21 += (Fx_21*dt)/m_2
    v_22 += (Fx_22*dt)/m_2
    v_23 += (Fx_23*dt)/m_2
    
    x_21 += v_21*dt
    x_22 += v_22*dt
    x_23 += v_23*dt
    
    t += dt
    
    x_11list.append(x_11)
    x_12list.append(x_12)

    x_21list.append(x_21)
    x_22list.append(x_22)
    
    v_21list.append(v_21)
    v_22list.append(v_22)
    
    Fx_21list.append(Fx_21)
    Fx_22list.append(Fx_22)
    
# Vizualisation
plt.figure(figsize=(10,10))
plt.plot(x_11list, x_12list, linewidth=2.0, label="Sun", color="darkorange")
plt.plot(x_21list, x_22list, linewidth=2.0, label="Earth", color="royalblue")
plt.quiver(x_21list[0], x_22list[0], v_21list[0], v_22list[0], label="Earth velocity vector", color='red')
plt.quiver(x_21list[0], x_22list[0], Fx_21list[0], Fx_22list[0], label=r"$\bf{F}_{21}$", color='green')
plt.xlabel(r"$x_m$")
plt.ylabel(r"$y_m$")
plt.title("Numerical Simulation of Newton's Law of Universal Gravitation")
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion

While Newton was able to formulate his law of gravity in his monumental work, he was deeply uncomfortable with the notion of "action at a distance" that his equations implied. In 1692, in his third letter to Bentley, he wrote: "That one body may act upon another at a distance through a vacuum without the mediation of anything else, by and through which their action and force may be conveyed from one another, is to me so great an absurdity that, I believe, no man who has in philosophic matters a competent faculty of thinking could ever fall into it."

He never, in his words, "assigned the cause of this power". In all other cases, he used the phenomenon of motion to explain the origin of various forces acting on bodies, but in the case of gravity, he was unable to experimentally identify the motion that produces the force of gravity (although he invented two mechanical hypotheses in 1675 and 1717).

These objections were explained by Einstein's theory of general relativity, in which gravitation is an attribute of curved spacetime instead of being due to a force propagated between bodies. In Einstein's theory, energy and momentum distort spacetime in their vicinity, and other particles move in trajectories determined by the geometry of spacetime. This allowed a description of the motions of light and mass that was consistent with all available observations. In general relativity, the gravitational force is a fictitious force due to the curvature of spacetime, because the gravitational acceleration of a body in free fall is due to its world line being a geodesic of spacetime [**[5]**](https://en.wikipedia.org/wiki/Newton%27s_law_of_universal_gravitation).
